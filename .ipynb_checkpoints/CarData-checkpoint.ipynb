```python
df = df.drop(columns=['Car_Name'])
df
```

```python
index_SP = df[df['Selling_Price'] > 25].index
index_kmsd = df[df['Kms_Driven'] > 300000].index

df_new1 = df.drop(index_SP)
df_new = df_new1.drop(index_kmsd)
df_new.set_index(pd.Series(range(0, 298)), inplace=True)
# print(index_SP)
# print(index_kmsd)
```

```python
mean = np.mean (df_new['Kms_Driven'])
meany = np.mean (df_new['Selling_Price'])

plt.figure(figsize=(10,8))
plt.scatter(df_new['Kms_Driven'], df_new['Selling_Price'], s=25, alpha=.5)
plt.plot((mean , mean), (0, max(df_new['Selling_Price'])), 'red')
plt.plot((0, max(df['Kms_Driven'])), (meany, meany), 'green')
plt.title('scatter Kms_Driven and target for data', fontsize=15)
plt.xlabel("Kms_Driven", fontsize=10)
plt.ylabel("Selling_Price", fontsize=10)
plt.grid()
plt.show()
print(mean)
print(meany)
```

```python
LE = LabelEncoder()

df_label = LE.fit_transform(df_new['Fuel_Type'])
df_label1 = LE.fit_transform(df_new['Seller_Type'])
df_label2 = LE.fit_transform(df_new['Transmission'])

df_label = pd.DataFrame(df_label, columns=['Fuel_Type_le'])
df_label1 = pd.DataFrame(df_label1, columns=['Seller_Type_le'])
df_label2 = pd.DataFrame(df_label2, columns=['Transmission_le'])

df_onehotencoding = pd.concat([df_new, df_label, df_label1, df_label2], axis=1)
df_processing = df_onehotencoding.drop(columns=['Fuel_Type', 'Seller_Type', 'Transmission'])
df_processing
```

```python
corr_matrix = df_processing.corr()
corr_matrix['Transmission_le'].sort_values(ascending=False)
```

```python
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation matrix of Breast Cancer data")
plt.show()
```

```python
scale = MinMaxScaler(feature_range=(0, 1))
norm1 = scale.fit_transform(df_processing[['Year','Selling_Price','Present_Price','Kms_Driven','Owner','Age']])
norm1 = pd.DataFrame(norm1, columns=['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Owner', 'Age'])
df_filnal = pd.concat([norm1, df_label, df_label1, df_label2], axis=1)
df_filnal 
```

```python
X_one = df_filnal['Present_Price'].values.reshape(-1, 1)
Y_one = df_filnal['Selling_Price'].values.reshape(-1, 1)
```

```python
x_train_one, x_test_one, y_train_one, y_test_one = train_test_split(X_one, Y_one, test_size=.2, random_state=0)
```

```python
print(x_train_one.shape)
print(x_test_one.shape)
print(y_train_one.shape)
print(y_test_one.shape)
```

```python
regressor_one = LinearRegression()
```

```python
regressor_one.fit(x_train_one, y_train_one)
```

```python
y_pred_one = regressor_one.predict(x_test_one)
```

```python
y_pred_one
```

```python
print(metrics.mean_absolute_error(y_test_one, y_pred_one))
print(metrics.mean_squared_error(y_test_one, y_pred_one))
print(metrics.r2_score(y_test_one, y_pred_one))
```
